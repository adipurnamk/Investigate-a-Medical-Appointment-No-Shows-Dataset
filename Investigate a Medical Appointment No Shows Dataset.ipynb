### Binary-Value Columns and no_snow Column
>Before exploring the binary-value columns, we need to make a crosstab function.
This function will compute the frequency for each variable.

These binary-value columns are:
* gender
* scholarship 
* hipertension
* diabetes
* alcoholism
* handcap
* sms_received

```python
def crosstab(index, column=df['no_show'], dataframe=df, sorter=False, ascending=False):
    """
    Compute a simple cross tabulation of two (or more) factors. 
    By default computes a frequency table of the factors 
    unless an array of values and an aggregation function are passed.
    https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.crosstab.html
    """
    df_cross = pd.crosstab(index, column).reset_index()

    df_cross['no_show_rate'] = df_cross['Yes'] / \
        (df_cross['No'] + df_cross['Yes']) * 100
    
    if sorter:
        df_cross.sort_values(by=['no_show_rate'], ignore_index=True, ascending=False, inplace=True)
        return df_cross
    else:
        return df_cross
```

>In this step we need to contain our binary-value columns into some container, such as a list.
Then we iterate through that container while applying crosstab function on it.

```python
# Make list containing binary-value columns
# 
binary_feature = ['gender', 'scholarship', 'hipertension', 'diabetes', 'alcoholism', 'handcap', 'sms_received']

# Iterate through list above, then add the absolute difference no_show value
# 
for b in binary_feature:
    df_cross = crosstab(index=df[b], column=df.no_show)
    diff = df_cross.iloc[0][3] - df_cross.iloc[1][3]
    display(df_cross)
    print(b, 'difference rate= ', abs(diff), '% \n')
```

> The difference rate is so small, for gender, scholarship, hipertension, diabetes, alcoholism, and handcap is no more than 5%. So we cannot use these variables for no_show deciding factor.

> Except for the sms_received column which the difference rate is more than 10%. Maybe we can use this as a deciding factor. Although it's kinda weird, the patient who received SMS showing that more no_show rate than non received SMS.

### Time Span
> In Data Cleaning section we has added the time span column. Then we apply the crosstab function to this column.
<li><a href="#add">Time Span Calculation</a></li>

```python
# Applying crosstab function to time span value, assign it into new data frame, df_time
#
df_time = crosstab(df['time_span'], column=df['no_show'], dataframe=df, sorter=False)
df_time.head()
```

```python
df_time.time_span.corr(df_time.no_show_rate)
```

> It's kind of difficult to decide wheter this variable is useful or not for predict no showing patient.
We need to group our data by cutting the data frame, like user [tsilveira](https://www.kaggle.com/tsilveira) did in this [article](https://www.kaggle.com/tsilveira/applying-heatmaps-for-categorical-data-analysis).

```python
# First we need to determine the grouping category for labels
# 
categories = pd.Series(['Same day: 0', 
                        'Short: 1-3', 
                        'Week: 4-7', 
                        'Fortnight: 8-15', 
                        'Month: 16-30', 
                        'Quarter: 31-90', 
                        'Semester: 91-180', 
                        'Very long: >180'])

# Apply .cut() method on time_span column, bins is bottom value for the category above,
# 
df_time['time_span'] = pd.cut(df_time.time_span, bins=[-1,0,3,7,15,30,90,180, 10000], labels=categories)

df_time_cut = df_time.groupby('time_span').sum().reset_index()
df_time_cut['no_show_rate'] = df_time_cut['Yes'] / (df_time_cut['No'] + df_time_cut['Yes']) * 100
df_time_cut
```

```python
# Since index 7 have empty value, we can drop it.
# 
df_time_cut.drop(7, inplace=True)
df_time_cut
```

>Then we plot the df_time_cut into bar plot.

```python
g_time_span = sns.barplot(data = df_time_cut, y = 'no_show_rate', x='time_span', color='royalblue')
g_time_span.set_xticklabels(g_time_span.get_xticklabels(), rotation=90);
```

> For the same day waiting time the no show rate is so small, it' just 6.63%. Then this rate is jumped high when the waiting time is not at the same day. Starting from short term, the rate is 23.6%. This value is increasing along as waiting time increased. This trend happened until quarter term, which is the highest no show rate for time span variable, 33.11%. At the semester term the no showing rate is decreased, it's just 27.3%.

### Appointment Weekday
>Same as before, we need to apply crosstab function on the appointment_day column. 

```python
df_weekday = crosstab(index=df['appointment_weekday'], column=df.no_show, dataframe=df, sorter=False)
df_weekday.head()
```

>We need to sort the row by weekdays, so we use .reindex() with cats list to sort it.

```python
# Make sort template
# 
cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Applying the method
# 
df_weekday = df_weekday.groupby(['appointment_weekday']).sum().reindex(cats).reset_index()
df_weekday
```

>Then we plot it into bar plot

```python
g_weekday = sns.barplot(data=df_weekday, y='no_show_rate', x='appointment_weekday', color='royalblue')
g_weekday.set_xticklabels(g_weekday.get_xticklabels(), rotation=90);
```

> From the barplot above, the highest no showing rate is happening on Saturday which is 23.08%. Actually, the trend is increased from Friday.  
> There is no appointment that happened on Sunday, we can know that health services on Sundays are closed.  
> Starting Sunday to Thursday the no showing trend is decreasing along the weekdays.  
> People tend to not coming to the health center in the weekend, or a day before the weekend.  
> From this insight, we can take appointment_weekdays as variable to determine no showing patient rate.
